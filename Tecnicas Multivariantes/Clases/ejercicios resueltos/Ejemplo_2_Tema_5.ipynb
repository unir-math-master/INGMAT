```python
# Correlaci贸n entre columnas num茅ricas
# ==============================================================================

def tidy_corr_matrix(corr_mat):
    '''
    Funci贸n para convertir una matriz de correlaci贸n de pandas en formato tidy
    '''
    corr_mat = corr_mat.stack().reset_index()
    corr_mat.columns = ['variable_1','variable_2','r']
    corr_mat = corr_mat.loc[corr_mat['variable_1'] != corr_mat['variable_2'], :]
    corr_mat['abs_r'] = np.abs(corr_mat['r'])
    corr_mat = corr_mat.sort_values('abs_r', ascending=False)
    
    return(corr_mat)



corr_matrix = datos.select_dtypes(include=['float64', 'int']).corr(method='pearson')
tidy_corr_matrix(corr_matrix).head(10)
```

```python
# Divisi贸n de los datos en train y test
# ==============================================================================
X = datos[['tv', 'radio', 'periodico']]
y = datos['ventas']

X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        y.values.reshape(-1,1),
                                        train_size   = 0.8,
                                        random_state = 1234,
                                        shuffle      = True
                                    )
```

```python
# Creaci贸n del modelo utilizando matrices como en scikitlearn
# ==============================================================================
# A la matriz de predictores se le tiene que a帽adir una columna de 1s para el intercept del modelo
X_train = sm.add_constant(X_train, prepend=True)
modelo = sm.OLS(endog=y_train, exog=X_train,)
modelo = modelo.fit()
print(modelo.summary())
```

```python
#El modelo con todas las variables introducidas como predictores tiene un  2
#alto (0.894), es capaz de explicar el 89.4% de la variabilidad observada en las ventas.
#El p-value del modelo es significativo (1.01e-75) por lo que se puede aceptar que
#el modelo es mejor que lo esperado es por azar, al menos uno de los coeficientes parciales de regresi贸n es distinto de 0.

#Acorde al p-value obtenido para el coeficiente parcial de regresi贸n de periodico (0.723),
#esta variable no contribuye de forma significativa al modelo. Se entrena de nuevo el modelo,
#pero esta vez excluyendo el predictor periodico.

# Creaci贸n del modelo utilizando matrices
# ==============================================================================
# Se elimina la columna periodico del conjunto de train y test
X_train = X_train.drop(columns = 'periodico')
X_test  = X_test.drop(columns = 'periodico')

# A la matriz de predictores se le tiene que a帽adir una columna de 1s para el
# intercept del modelo
X_train = sm.add_constant(X_train, prepend=True)
modelo  = sm.OLS(endog=y_train, exog=X_train,)
modelo  = modelo.fit()
print(modelo.summary())
```

```python
# Intervalos de confianza para los coeficientes del modelo
# ==============================================================================
intervalos_ci = modelo.conf_int(alpha=0.05)
intervalos_ci.columns = ['2.5%', '97.5%']
intervalos_ci
```

```python
# Diagn贸stico errores (residuos) de las predicciones de entrenamiento
# ==============================================================================
y_train = y_train.flatten()
prediccion_train = modelo.predict(exog = X_train)
residuos_train   = prediccion_train - y_train
```
